```python
trainning_df = trainning_df.iloc[0:TRAINNING_SIZE,:].copy()

trainning_article_ls = list(trainning_df['title'])
trainning_highlight_ls = list(trainning_df['quote'])

del trainning_df
```

```python
df = pd.DataFrame(columns=['target_text','source_text'])
df['target_text'] = trainning_highlight_ls
df['source_text'] = ['summarize: '+item for item in trainning_article_ls]
```

```python
df=df.dropna()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, test_size=0.2)
train_df.shape, test_df.shape
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
MAX_EPOCHS = 8
```

```python
model.train(train_df=train_df,
            eval_df=test_df,
            source_max_token_len = 512, 
            target_max_token_len = 128,
            batch_size = 8,
            max_epochs = MAX_EPOCHS,
            use_gpu = True,
            outputdir = "outputs/2",
            save_only_last_epoch=True
            )
```
